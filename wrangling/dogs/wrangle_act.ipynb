We could say that Golden Retriever most popular (most tweeted) breed among "WeRateDog" channel subscribers. But lets check mean probability that those dogs were recognized correctly. 

```python
stats = merged_df.groupby('recognized_breed').agg({
    'confidence_level': 
    {
     'confidence_mean': np.mean
    }, 
     'retweeted':
    {
        'retweeted_mean': np.mean,
        'retweeted_sum': np.sum
    },
    'likes':
    {
        'likes_sum':np.sum,
        'likes_mean': np.mean
    },
    'recognized_breed': 'count'
});
```

```python
stats.sort_values(('confidence_level', 'confidence_mean'), ascending=False).head(5)
```
